#### Monthly Trends:
Let's see if stocks tend to be higher or lower during certain months. Calculate the mean stock price of IBM, sorted by months.
* In which months has IBM historically had a higher stock price (on average)

```python
ax=cc.groupby(cc.Date.dt.month)['StockPrice'].mean().plot(lw=4,title='Mean Stock price sorted by months',figsize=(10,6))
ibm.groupby(ibm.Date.dt.month)['StockPrice'].mean().plot(lw=4)
ge.groupby(ge.Date.dt.month)['StockPrice'].mean().plot(lw=4)
pg.groupby(pg.Date.dt.month)['StockPrice'].mean().plot(lw=4)
boeing.groupby(boeing.Date.dt.month)['StockPrice'].mean().plot(lw=4)
ax.legend(['CC','IBM','GE','PG','BOEING'])
ax.set_xlabel('Months')
ax.set_ylabel('Mean Stock Price')
ax.set_yticks((40,50,60,70,80,90,100,110,120,130,140,150,160))
```

IBM is high in the months Jan-May
